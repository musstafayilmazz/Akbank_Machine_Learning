Then I searched for suburb rank for melbourne from this website https://www.auhouseprices.com/suburb-ranking/VIC/house/ . I couldnt find 2016 dataset in given time. If I could I would be able to fill nan values about the prices with this median value of suburbs

Then I scrapped data from this website and created my own dataset of suburb ranking

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


url = "https://www.auhouseprices.com/suburb-ranking/VIC/house/"


response = requests.get(url)


if response.status_code == 200:

    soup = BeautifulSoup(response.text, 'html.parser')

    table = soup.find('table', class_='table table-rank')


    ranks = []
    suburbs = []
    median_prices = []


    for row in table.find_all('tr'):
        cells = row.find_all('td')
        if len(cells) >= 5:
            rank = cells[1].text.strip()
            suburb = cells[3].text.strip()
            median_price = cells[4].text.strip()

            # Check if the median price starts with a "$"
            if not median_price.startswith('$'):
                # If it doesn't, copy it to the suburb
                suburbs.append(median_price)
            else:
                suburbs.append(suburb)
            ranks.append(rank)
            median_prices.append(median_price)


    df = pd.DataFrame({'Rank': ranks, 'Suburb': suburbs, 'House Median Price': median_prices})


    df = df[df['Rank'] != 'Rank']

    df = df.reset_index(drop=True)


    print(df)
else:
    print("Failed to retrieve the web page.")


response.close()
```

```python
y = data_copy["Price"]
X = data_copy.drop("Price", axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso, LinearRegression, Ridge, ElasticNet
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

```python

models = {
    'Lasso': Lasso(),
    'LinearRegression': LinearRegression(),
    'Ridge': Ridge(),
    'ElasticNet': ElasticNet(),
    'KNeighborsRegressor': KNeighborsRegressor(),
    'RandomForestRegressor': RandomForestRegressor(),
    'GradientBoostingRegressor': GradientBoostingRegressor(),
    'AdaBoostRegressor': AdaBoostRegressor(n_estimators=5, learning_rate=1.2, loss='exponential', random_state=2)
}


from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np


results = {
    'Model': [],
    'MAE': [],
    'MSE': [],
    'RMSE': [],
    'R2': []
}


for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)

    results['Model'].append(model_name)
    results['MAE'].append(mae)
    results['MSE'].append(mse)
    results['RMSE'].append(rmse)
    results['R2'].append(r2)


results_df = pd.DataFrame(results)


print(results_df)
best_model = results_df.loc[results_df['RMSE'].idxmin()]
print(f"\nBest Performing Model (based on RMSE): {best_model['Model']}")
```
